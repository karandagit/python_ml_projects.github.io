```python
import bs4
import urllib.request as url
```

```python
path = "https://www.linkedin.com/learning/"
```

```python
res = url.urlopen(path)
```

```python
page = bs4.BeautifulSoup(res)
```

```python
viewers = page.find_all('p',class_='learning-related-content-card__viewers')
```

```python
len(viewers)
```

```python
duration = page.find_all('div',class_='related-entity-media__duration')
```

```python
len(duration)
```

```python
titles = page.find_all('h3',class_='related-content-card__title')
```

```python
len(titles) 
```

```python
for div in page.find_all('div',class_='related-entity-media__duration'):
    print(div.text)
```

```python
for i in range(len(viewers)):
    print("Title :", titles[i].text)
    print("Duration :",duration[i].text)
    print("Viewers :",viewers[i].text)
    print("========================")
```

```python
data = {"titles":[],"viewers":[]}
for i in range(len(viewers)):
    data["titles"].append(titles[i].text)
    data["viewers"].append(viewers[i].text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```
